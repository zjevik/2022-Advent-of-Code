```python
# Input variable moves defined below for readability


###### Step 1
# Build directory as a dictionary. 
# Each directory title is a key with it's value as a list of sub directories 
# followed by the total size of all files (not including sub files)

def make_dir(_files, _current_dir):
    """Given a list of files, format them into a list of the form
    [sub_dir1, sub_dir2, ..., sub_dir_n, total other file size]"""
    size = 0
    result = []
    for file in _files:
        if file.startswith('dir'):
            result.append(f'{_current_dir}-{file.split(" ")[-1]}')
        else:
            size += int(file.split(' ')[0])
    result.append(size)

    return result

current_dir = ' '
current_files = []
dir_dict = dict()
# Remove useless list commands
lines = [x for x in lines if x != '$ ls']
for line in lines:
    if line.startswith('$ cd'): # Append last files and change directory
        if current_files:
            dir_dict[current_dir] = make_dir(current_files, current_dir)
            current_files = []
        if line.endswith('..'): # remove last dir
            current_dir = ('-').join(current_dir.split('-')[:-1])
        else:
            next_dir = line.split(' ')[-1]
            current_dir = f'{current_dir}-{next_dir}'
    else:
        current_files.append(line)

# Final time for last set of files
if current_files:
    dir_dict[current_dir] = make_dir(current_files, current_dir)
    current_files = []   
              
```

```python
###### Step 2    
    
# Make a dictionary of the total file sizes        
def total_size(_dir):
    """Recursively queries the subdirectories"""
    sub_dirs = dir_dict[_dir]
    if len(sub_dirs) == 1:
        return sub_dirs[0]
    else:
        return sum([total_size(sub_dir) for sub_dir in sub_dirs[:-1]]) + sub_dirs[-1]

dir_sizes = dict()
for key in dir_dict.keys():
    dir_sizes[key] = total_size(key)
```

```python
# Part 1    
total = 0
for val in dir_sizes.values():
    if val <= 100000:
        total += val   
print(total)     
```
