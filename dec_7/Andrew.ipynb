```python
class node:
    type = None
    content = None
    name = ''
    size = None
    cum_size = None
    parent = None
    def __init__(self, type, name,size=None, parent=None) -> None:
        self.content = []
        self.type = type
        self.name = name
        self.size = size
        self.parent = parent
        self.cum_size = 0

def pretty_str(root,level):
    space = ''.join(['  ' for _ in range(level)])
    print(f'{space}{root.name} ({root.type} size={root.size})')
    for child in root.content:
        pretty_str(child,level+1)
```

```python
root = node('dir','/')

current = None
last_command = ''
for l in lines:
    # print(f'line: {l}')
    l = l.replace('\n','')
    if l[0] == '$':
        last_command = l[2:4]
        if l[2:4] == 'cd':
            if l[5:] == '/':
                current = root
                continue
            elif l[5:] == '..':
                current = current.parent
                continue
            else:
                possible_dirs = [x for x in current.content if x.name == l[5:] and x.type == 'dir']
                if len(possible_dirs) > 1:
                    print(f'too many dirs with name "{l[5:]}"!!')
                if possible_dirs:
                    new = possible_dirs[0]
                else:
                    # print(f'making {l[5:]} in {current.name}')
                    new = node('dir',l[5:],parent=current)
                    current.content.append(new)
                current = new
                # print(f'current is {current.name}')
                continue
        else:
            #ls
            pass
    else:
        #ls
        t, name = l.split(' ')
        if t.isdigit():
            #file
            print(f'file {name}')
            matches = [x for x in current.content if x.name == name and x.type == 'file']
            if not matches:
                current.content.append(node('file',name,size=int(t),parent=current))
            else:
                print(f'file {name} already present')
        else:
            #dir
            print(f'dir: {name}')
            possible_dirs = [x for x in current.content if x.name == name and x.type == 'dir']
            if not possible_dirs:
                current.content.append(node('dir',name,parent=current))
        print(l)
    
# pretty_str(root,0)
```

```python
ans = []
def fill_sizes(root):
    size = 0
    possible_dirs = [x for x in root.content if x.type == 'dir']
    if possible_dirs:
        for dir in possible_dirs:
            size += fill_sizes(dir)
    files = [x for x in root.content if x.type == 'file']
    for file in files:
        size += file.size
    root.cum_size = size

    if size <= 100000:
        ans.append(size)

    # print(f'size of {root.name} is {size}')
    return size
fill_sizes(root)

sum(ans)
```
